```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#sample dataset
X = np.array([1,2,3,4,5])
y = np.array([2,4,5,4,5])
```

```python
#Initialize parameters
m,c = 0, 0
alpha = 0.01
ephocs = 1000
```

#Manually execute the LinearRegression

```python
#Gradient Descent
for _ in range(ephocs):
    y_pred = m*X+c
    error = y - y_pred
    m -= alpha*(-2*np.mean(X*error))
    c -= alpha*(-2*np.mean(error))
print("Slope (m):",round(m,3))
print("Intercept: (c): ",round(c,3))
```

#Part 2: Visualize of Fit

```python
plt.scatter(X,y, color="blue", label="Actual Data")
plt.plot(X, m*X+c, color='red', label="mx+c line")
plt.title("Manual Linear Regression Fit")
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```

```python
print("Question:\nScatter Points + regression line = ?")
print("Answer: Linear Regression Plot")
```

#Execute the LinearRegression using sklearn

```python
from sklearn.linear_model import LinearRegression
```

```python
#reshape for sklearn
X = X.reshape(-1, 1)
```

```python
#model
model = LinearRegression()
model.fit(X,y)
```

```python
print("Print Slop, Intercept and the prediction of a single input:")
print("Slope (m):",model.coef_[0])
print("Intercept (c):", model.intercept_)
print("Prediction for X=6: ",model.predict([[6]]))
```

```python
#Prediction
y_pred = model.predict(X)
```

```python
# Visualization
plt.scatter(X, y, color="blue", label="Actual Data")
plt.plot(X, y_pred, color="green", linewidth=2, label="Sklearn Regression Line")

plt.title("Sklearn Linear Regression Fit")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.show()
```

```python
#Evaluation
from sklearn.metrics import mean_squared_error ,r2_score

mse = mean_squared_error(y, y_pred)
print("MSE: ",round(mse,3))
score = r2_score(y, y_pred)
print("R² Score:", round(score, 3))
```
