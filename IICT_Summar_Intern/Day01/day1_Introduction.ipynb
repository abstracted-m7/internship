```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
arr = np.array([[1,2,3],[4,5,6]])
arr.dtype
```

```python
X = arr[0]
y = arr[1]
```

```python
print("X: ",X)
print("Y: ",y)
```

```python
plt.plot(X,y)
plt.show()
```

```python
plt.bar(X,y)
plt.xlim(0,10)
plt.ylim(0,10)
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.show()
```

```python
plt.scatter(X, y)
plt.show()
```
